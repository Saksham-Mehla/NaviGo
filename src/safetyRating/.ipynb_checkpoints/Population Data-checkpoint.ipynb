```python
import pandas as pd
```

```python
data = pd.read_excel("Datasets/2011-IndiaStateDist-0000.xlsx")
raw_df = pd.DataFrame(data, columns = ['State', 'District', 'Name', 'TOT_P'])
raw_df.head(10)
```

```python
df = pd.DataFrame(columns = ['State_Id', 'District_Id', 'State', 'District', 'Population'])

state = ''
for i, row in raw_df.iterrows():
    if row['State'] != 0:
        if row['District'] == 0:
            state = row['Name']
        else:
            if i%3==0:
                df2 = {'State_Id': row['State'], 'District_Id': row['District'], 'State': state, 'District': row['Name'], 'Population': row['TOT_P']}
                df = df.append(df2, ignore_index=True)
                
df
```

```python
df.to_csv('Datasets/dist_wise_population.csv')
```
